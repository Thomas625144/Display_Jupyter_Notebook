# Conclusions

Premièrement, nous avons constaté que l'insécurité alimentaire a continué de progresser (3%) sur les dernières années et représente en 2017 plus de 7%. Cependant, nous avons également constaté que la Terre possède assez de ressources pour nourrir 10,5 Milliards d'être-humains ou 8,6 Milliards d'être-humains si on prend en compte uniquement les ressources végétale.

En analysant l'adéquation entre les pays touchant le plus l'aide alimentaire et ceux étant le plus touché par l'insécurité alimentaire (point 6. E. b.) la conclusion va dans le sens de la mauvaise utilisation de l'aide alimentaire.

**On peut donc en conclure que les ressources mondiales sont mal réparties et/ou mal utilisées au regard de ses capacités et des besoins** Cela se confirme en étudiant le cas du Manioc en Thailande, qui montre une utilisation totalement inadaptée qui ne tient pas en compte des 6 Millions de personnes en insécurité alimentaire du pays.

**À la vue de cette étude, nous pouvons recommander à l'O.N.U de travailler sur la bonne répartion des ressources alimentaires en limitant les exportations de nourriture pour les pays possèdant une part importante de sa population en état de sous-nutrition.** L'objectif serait pour chaque pays de se concentrer sur sa population en insécurité alimentaire avant d'envisager l'exportation.
**Il serait aussi judicieux de limiter la part de l'alimentation animale qui n'est pas optimale car l'alimentation animale nécéssite plus de ressouces végétales qu'elle ne produit de ressouces animale.**
